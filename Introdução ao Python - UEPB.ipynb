# Métodos dos mínimos quadrados:

## Objetivos:

### Determinar os parâmetros de uma função;
### Obter o melhor ajuste para um conjunto de dados tentando minimizar a soma dos quadrados das diferenças entre o valor estimado e os dados observados.

# Equação utilizada:

# $ y= Ax + B $

# Pelo princípio dos mínimos quadrados, para minimizar o erro:

# $ T(A,B)=\sum(y-Ax-B)^2$

# Derivando T em relação a A e em relação a B:

 \begin{equation}
 \dfrac{\partial T}{\partial A}=2\sum(y-Ax-B)(-x)
 \end{equation}

\begin{equation}
 \dfrac{\partial T}{\partial B}=2\sum(y-Ax-B)(-1)
 \end{equation}

# Igualando a zero


## Para A

\begin{equation}
 \dfrac{\partial T}{\partial A}=-2\sum x(y-Ax-B)=0
 \end{equation}

\begin{equation}
 \sum xy-\sum Ax^2-\sum Bx=0
 \end{equation}

## Para B

\begin{equation}
 \dfrac{\partial T}{\partial B}=-2\sum(y-Ax-B)=0
 \end{equation}

\begin{equation}
 \sum y-\sum Ax-\sum B=0
 \end{equation}

# Montando o sistema

\begin{cases}
\sum xy = \sum Ax^2+\sum Bx  \\
 \sum y\ = \sum Ax+\sum B
\end{cases}

# Resolvendo o sistema:


\begin{array}{rrr}
\sum xy = & \sum x^2 & \sum x & A \\
\sum y = & \sum x & \sum 1  & B
\end{array}

# Matriz M (comum)

\begin{array}{rrr}
& \sum x^2 & \sum x \\
& \sum x & \sum 1  
\end{array}

# Matriz A

\begin{array}{rrr}
\sum xy & \sum x \\
\sum y & \sum 1 
\end{array}

# Matriz B

\begin{array}{rrr}
\sum x^2 & \sum xy \\
\sum x & \sum y
\end{array}

# Para determinar A:

\begin{equation}
 \frac{detMA}{detMM}
\end{equation}

# Para determinar B:

\begin{equation}
 \frac{detMB}{detMM}
\end{equation}

# Implementação no Python:

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

# Gerando uma série de dados:

```python
yx=[20,40,60,80,100]+np.random.standard_normal(5)*5
xx=[1,2,3,4,5]
plt.plot(xx,yx,'*')
```

# Matriz 

\begin{array}{rrr}
\sum xy = & \sum x^2 & \sum x & A \\
\sum y = & \sum x & \sum 1  & B
\end{array}

```python
# Montando os termos da matriz com zero:
M11=np.zeros(5) #usando numpy
M12=np.zeros(5)
M22=np.zeros(5)
# o termo M21=M12
V1=np.zeros(5) # termos antes da igualdade
V2=np.zeros(5)

print(M11)
print(M12)
```

# MATRIZ M
\begin{array}{rrr}
& \sum x^2 & \sum x \\
& \sum x & \sum 1  
\end{array}

```python
# Montando a matriz M:
for i in range(5):
    M11[i]=(xx[i])**2
    M12[i]=xx[i]
    M22[i]=1
# O TERMO M21=M12 não é preciso repetir
```

```python
# definindo os termos do lado esquerdo do sistema
for i in range(5):
    V1[i]=yx[i]*xx[i]
    V2[i]=yx[i]
```

# Montando a matriz com os termos


```python
MM=[[0,0],[0,0]]

MM=[[math.fsum(M11),math.fsum(M12)],[math.fsum(M12),math.fsum(M22)]]
print(MM)
```

# susbtituino o lado esquerdo do sistem na matriz

```python
MA=[[math.fsum(V1),math.fsum(M12)],[math.fsum(V2),math.fsum(M22)]] 
MB=[[math.fsum(M11),math.fsum(V1)],[math.fsum(M12),math.fsum(V2)]]
```

```python
# Calculando o determinante

detMM=np.linalg.det(MM) # função do numpy que calcula o determinante de uma matriz
detMA=np.linalg.det(MA)
detMB=np.linalg.det(MB)
```

```python
#Calculando os coeficientes A e B 
AA=detMA/detMM
BB=detMB/detMM
print(AA,BB)
```

```python
#Recuperando o ajuste
xx2=np.arange(0.5,5.5,0.1) #função para gerar a série np.arange(start,stop,step)
#xx2?
print(xx2)
Yx2=[0]*50
print(Yx2)
for j in range(50):
    Yx2[j]=AA*xx2[j]+BB
print(Yx2)
```

```python
#Plotting the results
plt.ylabel('YX')
plt.xlabel('XX')
plt.plot(xx,yx, 'k*')
plt.plot(xx2,Yx2, '-')
```

- Data Set:
 
 -- Esses dados podem ser tabulares (na forma de uma tabela), como os encontrados em bancos de dados relacionais, ou ainda podem ser coleções de arquivos ou documentos em formatos variados.

```python
from IPython.display import Image
Image("Array-List.jpg")
```

- Array:

    -- Um array em Python é uma coleção de elementos que têm o mesmo tipo de dado. Assim como a lista, um array é um objeto mutável e usa um índice para ordenar seus elementos.

    -- Uma lista em Python é uma coleção ordenada de objetos que pode conter elementos repetidos.
```python
A = np.array([[1,2,-3],[3,-1,2],[2,1,1]])
b = np.array([[1,0,2]])

# Multiplicação b*A
mult = np.dot(b, A)
print(f'b*A = {mult}.\n')

# Solução do sistema Ax = b
x = np.linalg.solve(A, b.T)
print(f'A solução do sistema Ax = b é:\n{x}.\n')

# Determinação dos autovetores e autovalores da matriz A
eig = np.linalg.eigh(A)
print(f'Autovalores: {eig[0]}.')
print(f'Autovetores:\n{eig[1]}.')
```

```python
species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```
